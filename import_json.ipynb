# Ancienne méthode

```python
def import_json(filename):
    with open(filename, 'r') as infile:
        text = infile.read()
        new_data = text.replace("\n","").replace('}{', '},{')
        json_data = json.loads(f'[{new_data}]')
    data = pd.json_normalize(json_data)
    return data
```
